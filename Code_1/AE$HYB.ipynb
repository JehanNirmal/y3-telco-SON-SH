```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, Multiply
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report, roc_auc_score
from sklearn.metrics import f1_score, matthews_corrcoef, precision_recall_curve
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedShuffleSplit, GridSearchCV
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE, ADASYN
from collections import Counter
import logging
import os
import gc
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Set random seed for reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Configuration
CONFIG = {
    'autoencoder': {
        'latent_dim': 8,
        'hidden_dims': [64, 32, 16],
        'activation': 'relu',
        'use_batch_norm': True,
        'reconstruction_weight': 0.7,
        'classification_weight': 0.3
    },
    'paper_autoencoder': {
        'hidden_dims': [12, 6, 6, 12],  # Paper's 20-12-6-6-12-20 structure
        'activation': 'relu',
        'dropout_rate': 0.5,
        'l1_reg': 0.01,
        'epochs': 100,
        'batch_size': 22  # Paper's batch size
    },
    'svm': {
        'kernel': ['rbf'],
        'C': [0.1, 1, 10],
        'gamma': ['scale', 'auto', 0.01]
    },
    'training': {
        'ae_epochs': 100,
        'batch_size': 32,
        'val_split': 0.2,
        'test_size': 0.1
    },
    'smote': {
        'N': 300,
        'k_neighbors': 4
    }
}

# ==================== DATA HANDLING ====================

def load_dataset(file_path: str) -> pd.DataFrame:
    """Load dataset from CSV file."""
    try:
        df = pd.read_csv(file_path)
        logger.info(f"Loaded dataset: {file_path} | Shape: {df.shape}")
        return df
    except Exception as e:
        logger.error(f"Failed to load {file_path}: {e}")
        raise

def preprocess_dataset(df: pd.DataFrame, test_size: float = 0.1) -> tuple[np.ndarray, np.ndarray, np.ndarray, np.ndarray, MinMaxScaler]:
    """Preprocess dataset: handle labels, drop irrelevant columns, normalize, and split."""
    label_candidates = [col for col in df.columns if col.lower() in ['label', 'labels']]
    if 'LABEL' not in df.columns and label_candidates:
        df = df.rename(columns={label_candidates[0]: 'LABEL'})
    elif 'LABEL' not in df.columns:
        raise ValueError("No label column found!")

    unique_labels = set(df['LABEL'].unique())
    if not unique_labels.issubset({0, 1}):
        df['LABEL'] = df['LABEL'].apply(lambda x: 1 if x == 3 else 0)

    drop_cols = [col for col in df.columns if any(x in col.lower() for x in ['userid', 'user_id', 'lat', 'lon', 'coord', 'timestamp', 'time'])]
    features = [col for col in df.columns if col not in drop_cols + ['LABEL']]
    X = df[features].values
    y = df['LABEL'].values

    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)

    sss = StratifiedShuffleSplit(n_splits=1, test_size=test_size, random_state=SEED)
    for train_idx, test_idx in sss.split(X_scaled, y):
        y_train, y_test = y[train_idx], y[test_idx]
        if len(np.unique(y_train)) < 2:
            sss2 = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
            for train_idx2, test_idx2 in sss2.split(X_scaled, y):
                X_train, X_test = X_scaled[train_idx2], X_scaled[test_idx2]
                y_train, y_test = y[train_idx2], y[test_idx2]
                break
        else:
            X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
        break

    logger.info(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
    logger.info(f"Train class distribution: {Counter(y_train)}")
    logger.info(f"Test class distribution: {Counter(y_test)}")
    return X_train, X_test, y_train, y_test, scaler

# ==================== PAPER'S AUTOENCODER IMPLEMENTATION ====================

class PaperAutoencoder:
    """Implementation of the autoencoder from the research paper."""
    
    def __init__(self, input_dim: int):
        self.input_dim = input_dim
        self.autoencoder = None
        self.encoder = None
        self.history = None
        self.reconstruction_threshold = None
        
    def build_model(self):
        """Build the autoencoder with paper's exact architecture."""
        config = CONFIG['paper_autoencoder']
        
        # Input layer
        input_layer = Input(shape=(self.input_dim,), name='input')
        
        # Apply L1 regularization at input layer (as mentioned in paper)
        x = Dense(self.input_dim, activation='linear', 
                 activity_regularizer=tf.keras.regularizers.l1(config['l1_reg']),
                 name='input_regularized')(input_layer)
        
        # Encoder layers: progressively smaller sizes
        encoded = x
        for i, dim in enumerate(config['hidden_dims'][:2]):  # First two layers: 12, 6
            encoded = Dense(dim, activation=config['activation'], 
                           name=f'encoder_{i+1}')(encoded)
            if i == 0:  # 50% dropout at first hidden layer
                encoded = Dropout(config['dropout_rate'])(encoded)
        
        # Decoder layers: mirror the encoder
        decoded = encoded
        for i, dim in enumerate(config['hidden_dims'][2:]):  # Last two layers: 6, 12
            decoded = Dense(dim, activation=config['activation'], 
                           name=f'decoder_{i+1}')(decoded)
        
        # Output layer: same size as input
        output = Dense(self.input_dim, activation='sigmoid', name='reconstruction')(decoded)
        
        # Create models
        self.autoencoder = Model(input_layer, output, name='paper_autoencoder')
        self.encoder = Model(input_layer, encoded, name='paper_encoder')
        
        # Compile with MSE loss (as mentioned in paper)
        self.autoencoder.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse',
            metrics=['mae']
        )
        
        logger.info(f"Paper Autoencoder built: {self.input_dim}-{'-'.join(map(str, config['hidden_dims']))}-{self.input_dim}")
        
    def train(self, X_train: np.ndarray, y_train: np.ndarray, 
              X_val: np.ndarray = None, y_val: np.ndarray = None):
        """Train the autoencoder using only normal (non-anomalous) data."""
        if self.autoencoder is None:
            self.build_model()
        
        config = CONFIG['paper_autoencoder']
        
        # Filter to only normal data (label = 0) for training
        normal_mask = y_train == 0
        X_normal = X_train[normal_mask]
        
        if len(X_normal) == 0:
            logger.warning("No normal samples found for training. Using all data.")
            X_normal = X_train
        
        logger.info(f"Training on {len(X_normal)} normal samples out of {len(X_train)} total")
        
        # Validation data (also only normal samples if provided)
        validation_data = None
        if X_val is not None and y_val is not None:
            val_normal_mask = y_val == 0
            X_val_normal = X_val[val_normal_mask]
            if len(X_val_normal) > 0:
                validation_data = (X_val_normal, X_val_normal)
        
        callbacks = [
            EarlyStopping(monitor='val_loss' if validation_data else 'loss', 
                         patience=20, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(monitor='val_loss' if validation_data else 'loss', 
                             factor=0.5, patience=10, min_lr=1e-6, verbose=1)
        ]
        
        # Train autoencoder to reconstruct normal data
        self.history = self.autoencoder.fit(
            X_normal, X_normal,  # Input and target are the same for autoencoder
            validation_data=validation_data,
            epochs=config['epochs'],
            batch_size=config['batch_size'],
            callbacks=callbacks,
            verbose=1
        )
        
        # Calculate reconstruction threshold on training data
        train_reconstructions = self.autoencoder.predict(X_normal, verbose=0)
        train_errors = np.mean(np.square(X_normal - train_reconstructions), axis=1)
        self.reconstruction_threshold = np.percentile(train_errors, 95)  # 95th percentile
        
        logger.info(f"Reconstruction threshold set to: {self.reconstruction_threshold:.6f}")
        
    def predict(self, X_test: np.ndarray) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
        """Predict anomalies based on reconstruction error."""
        if self.autoencoder is None:
            raise ValueError("Model must be trained before prediction")
        
        # Get reconstructions
        X_reconstructed = self.autoencoder.predict(X_test, verbose=0)
        
        # Calculate reconstruction errors
        reconstruction_errors = np.mean(np.square(X_test - X_reconstructed), axis=1)
        
        # Predict anomalies (high reconstruction error = anomaly)
        y_pred = (reconstruction_errors > self.reconstruction_threshold).astype(int)
        
        # Use reconstruction error as probability score (normalized)
        max_error = np.max(reconstruction_errors)
        y_proba = reconstruction_errors / (max_error + 1e-10)
        
        return y_pred, y_proba, reconstruction_errors
    
    def evaluate(self, X_test: np.ndarray, y_test: np.ndarray) -> dict:
        """Evaluate the autoencoder model."""
        y_pred, y_proba, reconstruction_errors = self.predict(X_test)
        
        # Optimize threshold using F1 score
        precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
        f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
        optimal_idx = np.argmax(f1_scores)
        optimal_threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else 0.5
        
        # Re-predict with optimal threshold
        y_pred_optimal = (y_proba > optimal_threshold).astype(int)
        
        roc_auc = roc_auc_score(y_test, y_proba) if len(np.unique(y_test)) > 1 else np.nan
        pr_auc = auc(recall, precision)
        
        return {
            'roc_auc': roc_auc,
            'pr_auc': pr_auc,
            'f1_score': f1_score(y_test, y_pred_optimal),
            'mcc': matthews_corrcoef(y_test, y_pred_optimal),
            'confusion_matrix': confusion_matrix(y_test, y_pred_optimal),
            'report': classification_report(y_test, y_pred_optimal, output_dict=True, zero_division=0),
            'y_pred': y_pred_optimal,
            'y_true': y_test,
            'y_proba': y_proba,
            'optimal_threshold': optimal_threshold,
            'reconstruction_errors': reconstruction_errors,
            'reconstruction_threshold': self.reconstruction_threshold
        }

# ==================== PLOTTING UTILITIES (FIXED) ====================

class ResultsPlotter:
    """Comprehensive plotting utilities for anomaly detection results."""
    
    def __init__(self, figsize=(12, 8)):
        self.figsize = figsize
        self.colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
        
    def plot_training_history(self, history, title="Training History", save_path=None):
        """Plot training history for autoencoder using Plotly - FIXED VERSION."""
        # Check if this is a supervised autoencoder (has classification metrics)
        is_supervised = any(key.startswith('classification_') for key in history.history.keys())
        
        if is_supervised:
            # Supervised autoencoder with multiple outputs
            fig = make_subplots(rows=2, cols=2, subplot_titles=(
                'Total Loss', 'Reconstruction Loss', 'Classification Loss', 'Classification Accuracy'
            ))
            
            # Total loss
            fig.add_trace(go.Scatter(y=history.history['loss'], name='Training Loss', 
                                    line=dict(color=self.colors[0])), row=1, col=1)
            if 'val_loss' in history.history:
                fig.add_trace(go.Scatter(y=history.history['val_loss'], name='Validation Loss', 
                                        line=dict(color=self.colors[1])), row=1, col=1)
            
            # Reconstruction loss
            fig.add_trace(go.Scatter(y=history.history['reconstruction_loss'], name='Training', 
                                    line=dict(color=self.colors[2])), row=1, col=2)
            if 'val_reconstruction_loss' in history.history:
                fig.add_trace(go.Scatter(y=history.history['val_reconstruction_loss'], name='Validation', 
                                        line=dict(color=self.colors[3])), row=1, col=2)
            
            # Classification loss
            fig.add_trace(go.Scatter(y=history.history['classification_loss'], name='Training', 
                                    line=dict(color=self.colors[4])), row=2, col=1)
            if 'val_classification_loss' in history.history:
                fig.add_trace(go.Scatter(y=history.history['val_classification_loss'], name='Validation', 
                                        line=dict(color=self.colors[5])), row=2, col=1)
            
            # Classification accuracy
            fig.add_trace(go.Scatter(y=history.history['classification_accuracy'], name='Training', 
                                    line=dict(color=self.colors[0])), row=2, col=2)
            if 'val_classification_accuracy' in history.history:
                fig.add_trace(go.Scatter(y=history.history['val_classification_accuracy'], name='Validation', 
                                        line=dict(color=self.colors[1])), row=2, col=2)
            
            fig.update_layout(title_text=title, height=600, width=800, showlegend=True)
            fig.update_xaxes(title_text="Epoch")
            fig.update_yaxes(title_text="Loss", row=1, col=1)
            fig.update_yaxes(title_text="MSE Loss", row=1, col=2)
            fig.update_yaxes(title_text="Binary Crossentropy", row=2, col=1)
            fig.update_yaxes(title_text="Accuracy", row=2, col=2)
            
        else:
            # Simple autoencoder (paper's autoencoder)
            fig = make_subplots(rows=1, cols=2, subplot_titles=('Training Loss', 'Training MAE'))
            
            fig.add_trace(go.Scatter(y=history.history['loss'], name='Training Loss', 
                                    line=dict(color=self.colors[0])), row=1, col=1)
            if 'val_loss' in history.history:
                fig.add_trace(go.Scatter(y=history.history['val_loss'], name='Validation Loss', 
                                        line=dict(color=self.colors[1])), row=1, col=1)
            
            # Use the correct MAE metric name - FIXED
            mae_key = 'mae' if 'mae' in history.history else 'reconstruction_mae'
            val_mae_key = 'val_mae' if 'val_mae' in history.history else 'val_reconstruction_mae'
            
            if mae_key in history.history:
                fig.add_trace(go.Scatter(y=history.history[mae_key], name='Training MAE', 
                                        line=dict(color=self.colors[2])), row=1, col=2)
            if val_mae_key in history.history:
                fig.add_trace(go.Scatter(y=history.history[val_mae_key], name='Validation MAE', 
                                        line=dict(color=self.colors[3])), row=1, col=2)
            
            fig.update_layout(title_text=title, height=400, width=800, showlegend=True)
            fig.update_xaxes(title_text="Epoch")
            fig.update_yaxes(title_text="MSE Loss", row=1, col=1)
            fig.update_yaxes(title_text="MAE", row=1, col=2)
        
        if save_path:
            fig.write_image(save_path, scale=2)
        fig.show()
        
    def plot_reconstruction_errors(self, reconstruction_errors, y_true, threshold, 
                                  title="Reconstruction Error Distribution", save_path=None):
        """Plot reconstruction error distribution."""
        fig = go.Figure()
        
        # Normal samples
        normal_errors = reconstruction_errors[y_true == 0]
        fig.add_trace(go.Histogram(x=normal_errors, name='Normal', opacity=0.7, 
                                  marker_color=self.colors[0], nbinsx=50))
        
        # Anomalous samples
        anomaly_errors = reconstruction_errors[y_true == 1]
        if len(anomaly_errors) > 0:
            fig.add_trace(go.Histogram(x=anomaly_errors, name='Anomaly', opacity=0.7, 
                                      marker_color=self.colors[1], nbinsx=50))
        
        # Threshold line
        fig.add_vline(x=threshold, line_dash="dash", line_color="red", 
                     annotation_text=f"Threshold: {threshold:.4f}")
        
        fig.update_layout(title=title, xaxis_title="Reconstruction Error", 
                         yaxis_title="Frequency", barmode='overlay')
        
        if save_path:
            fig.write_image(save_path, scale=2)
        fig.show()
        
    def plot_feature_visualization(self, X_original, X_extracted, y, title="Feature Visualization", save_path=None):
        """Visualize original vs extracted features using t-SNE."""
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        fig.suptitle(title, fontsize=16, fontweight='bold')
        
        # Ensure perplexity is valid for dataset size
        n_samples = len(X_original)
        perplexity = min(30, max(5, n_samples // 4 - 1))  # Adjust perplexity dynamically
        
        try:
            if X_original.shape[1] > 50:
                pca = PCA(n_components=50, random_state=SEED)
                X_original_reduced = pca.fit_transform(X_original)
                tsne_original = TSNE(n_components=2, random_state=SEED, perplexity=perplexity)
                X_original_2d = tsne_original.fit_transform(X_original_reduced)
            else:
                tsne_original = TSNE(n_components=2, random_state=SEED, perplexity=perplexity)
                X_original_2d = tsne_original.fit_transform(X_original)
        except Exception as e:
            logger.warning(f"t-SNE failed for original features: {e}. Skipping visualization.")
            axes[0].text(0.5, 0.5, 't-SNE Failed', ha='center', va='center')
            axes[0].set_title(f'Original Features (t-SNE)\n{X_original.shape[1]} dimensions')
            axes[0].grid(True, alpha=0.3)
            return
        
        try:
            tsne_extracted = TSNE(n_components=2, random_state=SEED, perplexity=perplexity)
            X_extracted_2d = tsne_extracted.fit_transform(X_extracted)
        except Exception as e:
            logger.warning(f"t-SNE failed for extracted features: {e}. Skipping visualization.")
            axes[1].text(0.5, 0.5, 't-SNE Failed', ha='center', va='center')
            axes[1].set_title(f'Extracted Features (t-SNE)\n{X_extracted.shape[1]} dimensions')
            axes[1].grid(True, alpha=0.3)
            return
        
        axes[0].scatter(X_original_2d[y==0, 0], X_original_2d[y==0, 1], c=self.colors[0], alpha=0.6, label='Normal', s=20)
        axes[0].scatter(X_original_2d[y==1, 0], X_original_2d[y==1, 1], c=self.colors[1], alpha=0.8, label='Anomaly', s=20)
        axes[0].set_title(f'Original Features (t-SNE)\n{X_original.shape[1]} dimensions')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        axes[1].scatter(X_extracted_2d[y==0, 0], X_extracted_2d[y==0, 1], c=self.colors[0], alpha=0.6, label='Normal', s=20)
        axes[1].scatter(X_extracted_2d[y==1, 0], X_extracted_2d[y==1, 1], c=self.colors[1], alpha=0.8, label='Anomaly', s=20)
        axes[1].set_title(f'Extracted Features (t-SNE)\n{X_extracted.shape[1]} dimensions')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
        plt.show()
        
    def plot_roc_curves(self, results_dict, dataset_name="Dataset", save_path=None):
        """Plot ROC curves for all methods using Plotly."""
        fig = make_subplots(rows=1, cols=2, subplot_titles=('Paper Autoencoder', 'Supervised Hybrid+ADASYN'))
        
        method_keys = ['paper_autoencoder', 'supervised_hybrid_adasyn']
        method_names = ['Paper Autoencoder', 'Supervised Hybrid+ADASYN']
        
        for idx, (method_key, method_name) in enumerate(zip(method_keys, method_names)):
            if method_key in results_dict and 'y_true' in results_dict[method_key]:
                y_true = results_dict[method_key]['y_true']
                y_proba = results_dict[method_key]['y_proba']
                
                if len(np.unique(y_true)) > 1:
                    fpr, tpr, _ = roc_curve(y_true, y_proba)
                    roc_auc = auc(fpr, tpr)
                    fig.add_trace(go.Scatter(x=fpr, y=tpr, name=f'{method_name} (AUC = {roc_auc:.4f})', 
                                            line=dict(color=self.colors[idx])), row=1, col=idx+1)
                    fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], line=dict(color='gray', dash='dash'), 
                                            showlegend=False), row=1, col=idx+1)
                else:
                    fig.add_annotation(text='Insufficient classes', x=0.5, y=0.5, showarrow=False, 
                                      row=1, col=idx+1)
            
            fig.update_xaxes(title_text="False Positive Rate", row=1, col=idx+1)
            fig.update_yaxes(title_text="True Positive Rate", row=1, col=idx+1)
        
        fig.update_layout(title_text=f'ROC Curves - {dataset_name}', height=400, width=800)
        if save_path:
            fig.write_image(save_path, scale=2)
        fig.show()

# ==================== SUPERVISED AUTOENCODER ====================

class SupervisedAutoencoderFeatureExtractor:
    """Enhanced supervised autoencoder with attention and plotting."""
    
    def __init__(self, input_dim: int, latent_dim: int = 8, hidden_dims: list = None, 
                 activation: str = 'relu', use_batch_norm: bool = True,
                 reconstruction_weight: float = 0.7, classification_weight: float = 0.3):
        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.hidden_dims = hidden_dims or [64, 32, 16]
        self.activation = activation
        self.use_batch_norm = use_batch_norm
        self.reconstruction_weight = reconstruction_weight
        self.classification_weight = classification_weight
        self.sigma_recon = tf.Variable(1.0, trainable=True, dtype=tf.float32)
        self.sigma_class = tf.Variable(1.0, trainable=True, dtype=tf.float32)
        self.autoencoder = None
        self.encoder = None
        self.classifier = None
        self.combined_model = None
        self.history = None
        self.plotter = ResultsPlotter()
        
    def build_model(self):
        """Build the supervised autoencoder with attention and batch normalization."""
        input_layer = Input(shape=(self.input_dim,), name='input')
        
        attention = Dense(self.input_dim, activation='softmax', name='attention')(input_layer)
        attended_input = Multiply()([input_layer, attention])
        
        encoded = attended_input
        for i, dim in enumerate(self.hidden_dims):
            encoded = Dense(dim, activation=self.activation, 
                           kernel_regularizer=tf.keras.regularizers.l2(0.01), 
                           name=f'encoder_{i+1}')(encoded)
            if self.use_batch_norm:
                encoded = BatchNormalization()(encoded)
            encoded = Dropout(0.2)(encoded)
        
        latent = Dense(self.latent_dim, activation=self.activation, name='latent')(encoded)
        
        decoded = latent
        for i, dim in enumerate(reversed(self.hidden_dims)):
            decoded = Dense(dim, activation=self.activation, 
                           kernel_regularizer=tf.keras.regularizers.l2(0.01), 
                           name=f'decoder_{i+1}')(decoded)
            if self.use_batch_norm:
                decoded = BatchNormalization()(decoded)
            decoded = Dropout(0.2)(decoded)
        
        reconstruction_output = Dense(self.input_dim, activation='sigmoid', name='reconstruction')(decoded)
        classification_branch = Dense(32, activation=self.activation, name='classifier_hidden')(latent)
        classification_branch = Dropout(0.3)(classification_branch)
        classification_output = Dense(1, activation='sigmoid', name='classification')(classification_branch)
        
        self.encoder = Model(input_layer, latent, name='encoder')
        self.autoencoder = Model(input_layer, reconstruction_output, name='autoencoder')
        self.classifier = Model(input_layer, classification_output, name='classifier')
        self.combined_model = Model(
            inputs=input_layer, 
            outputs=[reconstruction_output, classification_output],
            name='supervised_autoencoder'
        )
        
        logger.info(f"Supervised Autoencoder built: Input={self.input_dim}, Latent={self.latent_dim}")
        
    def train(self, X_train: np.ndarray, y_train: np.ndarray, 
              X_val: np.ndarray = None, y_val: np.ndarray = None,
              epochs: int = 100, batch_size: int = 32, plot_training: bool = True):
        """Train the supervised autoencoder with weighted loss."""
        if X_train.shape[0] < 2:
            raise ValueError("At least two samples are required for training")
        if X_train.shape[1] != self.input_dim:
            raise ValueError(f"Input dimension mismatch: expected {self.input_dim}, got {X_train.shape[1]}")
        
        if self.combined_model is None:
            self.build_model()
        
        # Calculate class weights for classification loss
        class_weights = {0: 1.0, 1: sum(y_train == 0) / (sum(y_train == 1) + 1e-10)}
        
        # Custom loss function for classification with class weights
        def weighted_binary_crossentropy(y_true, y_pred):
            bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
            weight_vector = tf.where(y_true > 0, tf.cast(class_weights[1], tf.float32), tf.cast(class_weights[0], tf.float32))
            weighted_bce = weight_vector * bce
            return tf.reduce_mean(weighted_bce) / (2 * self.sigma_class ** 2) + tf.math.log(self.sigma_class)
        
        # Compile with custom classification loss
        self.combined_model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss={
                'reconstruction': lambda y_true, y_pred: tf.reduce_mean(tf.square(y_true - y_pred)) / (2 * self.sigma_recon ** 2) + tf.math.log(self.sigma_recon),
                'classification': weighted_binary_crossentropy
            },
            loss_weights={
                'reconstruction': self.reconstruction_weight,
                'classification': self.classification_weight
            },
            metrics={
                'reconstruction': ['mae'],
                'classification': ['accuracy', 'precision', 'recall']
            }
        )
        
        train_targets = {
            'reconstruction': X_train,
            'classification': y_train
        }
        validation_data = None
        if X_val is not None and y_val is not None:
            validation_data = (X_val, {'reconstruction': X_val, 'classification': y_val})
        
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True, verbose=1),
            ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6, verbose=1)
        ]
        
        self.history = self.combined_model.fit(
            X_train, train_targets,
            validation_data=validation_data,
            epochs=epochs,
            batch_size=batch_size,
            callbacks=callbacks,
            verbose=1
        )
        
        if plot_training and self.history:
            self.plotter.plot_training_history(self.history, title=f"Supervised Autoencoder Training History\n({self.input_dim}→{self.latent_dim} features)")
        
    def extract_features(self, X: np.ndarray, batch_size: int = 1000) -> np.ndarray:
        """Extract features using the trained encoder in batches."""
        if self.encoder is None:
            raise ValueError("Model must be trained before feature extraction")
        
        features = []
        for i in range(0, len(X), batch_size):
            batch = X[i:i+batch_size]
            features.append(self.encoder.predict(batch, verbose=0))
        return np.vstack(features)
    
    def predict_classification(self, X: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
        """Get classification predictions and probabilities."""
        if self.classifier is None:
            raise ValueError("Model must be trained before prediction")
        
        y_proba = self.classifier.predict(X, verbose=0).flatten()
        y_pred = (y_proba > 0.5).astype(int)
        return y_pred, y_proba
    
    def get_reconstruction_error(self, X: np.ndarray) -> np.ndarray:
        """Calculate reconstruction error."""
        if self.autoencoder is None:
            raise ValueError("Model must be trained before reconstruction")
        
        X_reconstructed = self.autoencoder.predict(X, verbose=0)
        return np.mean(np.square(X - X_reconstructed), axis=1)

# ==================== SUPERVISED HYBRID ====================

class SupervisedHybridAESVM:
    """Enhanced supervised hybrid model combining autoencoder and SVM."""
    
    def __init__(self, input_dim: int, latent_dim: int = 8, hidden_dims: list = None,
                 activation: str = 'relu', use_batch_norm: bool = True,
                 reconstruction_weight: float = 0.7, classification_weight: float = 0.3):
        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.ae_extractor = SupervisedAutoencoderFeatureExtractor(
            input_dim, latent_dim, hidden_dims, activation, use_batch_norm,
            reconstruction_weight, classification_weight
        )
        self.svm_model = None
        self.scaler = MinMaxScaler()
        self.training_time = 0
        self.testing_time = 0
        self.plotter = ResultsPlotter()
        
    def train(self, X_train: np.ndarray, y_train: np.ndarray, 
              ae_epochs: int = 100, svm_param_grid: dict = None,
              plot_training: bool = True):
        """Train the hybrid model with ADASYN oversampling."""
        import time
        start_time = time.time()
        
        logger.info("Applying ADASYN to input data...")
        X_train_res, y_train_res = ADASYN(random_state=SEED).fit_resample(X_train, y_train)
        logger.info(f"After ADASYN: {Counter(y_train_res)}")
        
        val_split = CONFIG['training']['val_split']
        split_idx = int((1 - val_split) * len(X_train_res))
        X_ae_train, X_ae_val = X_train_res[:split_idx], X_train_res[split_idx:]
        y_ae_train, y_ae_val = y_train_res[:split_idx], y_train_res[split_idx:]
        
        self.ae_extractor.train(
            X_ae_train, y_ae_train, X_ae_val, y_ae_val,
            epochs=ae_epochs, batch_size=CONFIG['training']['batch_size'],
            plot_training=plot_training
        )
        
        X_train_features = self.ae_extractor.extract_features(X_train_res)
        X_train_features = self.scaler.fit_transform(X_train_features)
        
        if svm_param_grid is None:
            svm_param_grid = CONFIG['svm']
        self.svm_model = GridSearchCV(SVC(probability=True, random_state=SEED), svm_param_grid, cv=5)
        self.svm_model.fit(X_train_features, y_train_res)
        logger.info(f"Best SVM parameters: {self.svm_model.best_params_}")
        
        self.training_time = time.time() - start_time
        logger.info(f"Training completed in {self.training_time:.2f} seconds")
        
    def evaluate(self, X_test: np.ndarray, y_test: np.ndarray, 
                 plot_features: bool = True, dataset_name: str = "Dataset") -> dict:
        """Evaluate the hybrid model with optimized threshold."""
        import time
        start_time = time.time()
        
        X_test_features = self.ae_extractor.extract_features(X_test)
        X_test_features = self.scaler.transform(X_test_features)
        
        if plot_features:
            self.plotter.plot_feature_visualization(X_test, X_test_features, y_test, 
                                                   title=f"Feature Visualization - {dataset_name}")
        
        y_proba_svm = self.svm_model.predict_proba(X_test_features)[:, 1]
        precision, recall, thresholds = precision_recall_curve(y_test, y_proba_svm)
        f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
        optimal_threshold = thresholds[np.argmax(f1_scores)]
        y_pred_svm = (y_proba_svm > optimal_threshold).astype(int)
        
        y_pred_ae, y_proba_ae = self.ae_extractor.predict_classification(X_test)
        reconstruction_errors = self.ae_extractor.get_reconstruction_error(X_test)
        
        self.testing_time = time.time() - start_time
        
        result = {
            'roc_auc': roc_auc_score(y_test, y_proba_svm) if len(np.unique(y_test)) > 1 else np.nan,
            'pr_auc': auc(recall, precision),
            'f1_score': f1_score(y_test, y_pred_svm),
            'mcc': matthews_corrcoef(y_test, y_pred_svm),
            'confusion_matrix': confusion_matrix(y_test, y_pred_svm),
            'report': classification_report(y_test, y_pred_svm, output_dict=True, zero_division=0),
            'y_pred': y_pred_svm,
            'y_true': y_test,
            'y_proba': y_proba_svm,
            'optimal_threshold': optimal_threshold,
            'ae_roc_auc': roc_auc_score(y_test, y_proba_ae) if len(np.unique(y_test)) > 1 else np.nan,
            'ae_confusion_matrix': confusion_matrix(y_test, y_pred_ae),
            'ae_report': classification_report(y_test, y_pred_ae, output_dict=True, zero_division=0),
            'reconstruction_errors': reconstruction_errors,
            'training_time': self.training_time,
            'testing_time': self.testing_time,
            'n_features_original': self.input_dim,
            'n_features_extracted': self.latent_dim
        }
        return result

# ==================== MAIN EXECUTION ====================

def run_comparison(data_dir: str, dataset_files: list):
    """Run comparison between Paper's Autoencoder and Supervised Hybrid with ADASYN."""
    logger.info("="*80)
    logger.info("ANOMALY DETECTION COMPARISON: PAPER'S AUTOENCODER vs SUPERVISED HYBRID+ADASYN")
    logger.info("="*80)
    
    plotter = ResultsPlotter()
    all_results = {
        'paper_autoencoder': [],
        'supervised_hybrid_adasyn': []
    }
    dataset_results = []
    
    # Update dataset file paths with data_dir
    dataset_files = [(os.path.join(data_dir, os.path.basename(fp)), meta) for fp, meta in dataset_files]
    
    for i, (df, (file_path, _)) in enumerate(zip([load_dataset(fp) for fp, _ in dataset_files], dataset_files)):
        logger.info(f"\nProcessing Dataset {i+1}: {os.path.basename(file_path)}")
        
        X_train, X_test, y_train, y_test, scaler = preprocess_dataset(df, test_size=CONFIG['training']['test_size'])
        input_dim = X_train.shape[1]
        
        dataset_result = {}
        
        # Paper's Autoencoder
        logger.info("\n--- Paper's Autoencoder ---")
        paper_ae = PaperAutoencoder(input_dim)
        
        # Split training data for validation
        val_split = CONFIG['training']['val_split']
        split_idx = int((1 - val_split) * len(X_train))
        X_ae_train, X_ae_val = X_train[:split_idx], X_train[split_idx:]
        y_ae_train, y_ae_val = y_train[:split_idx], y_train[split_idx:]
        
        paper_ae.train(X_ae_train, y_ae_train, X_ae_val, y_ae_val)
        dataset_result['paper_autoencoder'] = paper_ae.evaluate(X_test, y_test)
        all_results['paper_autoencoder'].append(dataset_result['paper_autoencoder'])
        
        # Plot training history and reconstruction errors
        if paper_ae.history:
            plotter.plot_training_history(paper_ae.history, 
                                         title=f"Paper's Autoencoder Training - Dataset {i+1}")
        
        plotter.plot_reconstruction_errors(
            dataset_result['paper_autoencoder']['reconstruction_errors'],
            y_test,
            dataset_result['paper_autoencoder']['reconstruction_threshold'],
            title=f"Reconstruction Error Distribution - Dataset {i+1}"
        )
        
        # Supervised Hybrid + ADASYN
        logger.info("\n--- Supervised Hybrid + ADASYN ---")
        hybrid_model = SupervisedHybridAESVM(input_dim, **CONFIG['autoencoder'])
        hybrid_model.train(X_train, y_train, ae_epochs=CONFIG['training']['ae_epochs'])
        dataset_result['supervised_hybrid_adasyn'] = hybrid_model.evaluate(X_test, y_test, 
                                                                          dataset_name=f"Dataset {i+1}")
        all_results['supervised_hybrid_adasyn'].append(dataset_result['supervised_hybrid_adasyn'])
        
        dataset_results.append(dataset_result)
        plotter.plot_roc_curves(dataset_result, dataset_name=f"Dataset {i+1}")
        
        logger.info(f"\nDataset {i+1} Results:")
        logger.info(f"  Paper's Autoencoder: AUC={dataset_result['paper_autoencoder']['roc_auc']:.4f}, "
                    f"F1={dataset_result['paper_autoencoder']['f1_score']:.4f}, "
                    f"MCC={dataset_result['paper_autoencoder']['mcc']:.4f}")
        logger.info(f"  Supervised Hybrid+ADASYN: AUC={dataset_result['supervised_hybrid_adasyn']['roc_auc']:.4f}, "
                    f"F1={dataset_result['supervised_hybrid_adasyn']['f1_score']:.4f}, "
                    f"MCC={dataset_result['supervised_hybrid_adasyn']['mcc']:.4f}")
        
        gc.collect()
    
    logger.info("\n" + "="*80)
    logger.info("FINAL RESULTS SUMMARY")
    logger.info("="*80)
    
    print(f"\n{'Dataset':<10} {'Paper AE AUC':<12} {'Hybrid AUC':<12} {'Paper AE F1':<12} {'Hybrid F1':<12} {'Paper AE MCC':<12} {'Hybrid MCC':<12}")
    print("-" * 80)
    
    for i in range(len(dataset_files)):
        paper_ae = all_results['paper_autoencoder'][i]
        hybrid = all_results['supervised_hybrid_adasyn'][i]
        print(f"Dataset {i+1:<3} {paper_ae['roc_auc']:<12.4f} {hybrid['roc_auc']:<12.4f} "
              f"{paper_ae['f1_score']:<12.4f} {hybrid['f1_score']:<12.4f} "
              f"{paper_ae['mcc']:<12.4f} {hybrid['mcc']:<12.4f}")
    
    avg_results = {
        m: np.mean([r['roc_auc'] for r in all_results[m] if not np.isnan(r['roc_auc'])]) for m in all_results
    }
    avg_f1 = {
        m: np.mean([r['f1_score'] for r in all_results[m]]) for m in all_results
    }
    avg_mcc = {
        m: np.mean([r['mcc'] for r in all_results[m]]) for m in all_results
    }
    
    print("-" * 80)
    print(f"{'AVERAGE':<10} {avg_results['paper_autoencoder']:<12.4f} {avg_results['supervised_hybrid_adasyn']:<12.4f} "
          f"{avg_f1['paper_autoencoder']:<12.4f} {avg_f1['supervised_hybrid_adasyn']:<12.4f} "
          f"{avg_mcc['paper_autoencoder']:<12.4f} {avg_mcc['supervised_hybrid_adasyn']:<12.4f}")
    
    best_auc = max(avg_results.values())
    best_method = 'Supervised Hybrid+ADASYN' if avg_results['supervised_hybrid_adasyn'] == best_auc else "Paper's Autoencoder"
    logger.info(f"\nBest Method (by AUC): {best_method} (Average AUC: {best_auc:.4f})")

if __name__ == "__main__":
    # Configurable data directory
    data_dir = os.getenv('DATA_DIR', r'C:\Users\USER\Downloads\Sim-Dim-master\Sim-Dim-master')
    dataset_files = [
        ('dataset1.csv', 25),
        ('dataset2.csv', 25),
        ('dataset3.csv', 114),
        ('dataset4.csv', 114)
    ]
    run_comparison(data_dir, dataset_files)
```
