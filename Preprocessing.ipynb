```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\USER\Documents\Year 3\Research\Data\pm_data.csv")
```

```python
pm_data = pd.DataFrame(data)

# Get unique values in the 'cell_name_id' column
unique_values = pm_data["cell_name_id"].unique()

# Specify the directory to save files
save_dir = r"C:\Users\USER\Documents\Year 3\Research\Unique Cells"

# Ensure the directory exists
os.makedirs(save_dir, exist_ok=True)

# Save each filtered DataFrame to the specified directory
for value in unique_values:
    # Filter the DataFrame for each unique value
    df = pm_data[pm_data["cell_name_id"] == value]
    
    # Construct the full file path
    filename = os.path.join(save_dir, f"df_{value}.csv")
    
    # Save the DataFrame to a CSV file
    df.to_csv(filename, index=False)
    
    print(f"Saved {filename}")
```

```python
directory = r"C:\Users\USER\Documents\Year 3\Research\Unique Cells"

files = [f for f in os.listdir(directory) if f.endswith(".csv")]

for file in files:
    file_path = os.path.join(directory, file)
    
    df = pd.read_csv(file_path)
    
    null_summary = df.isnull().sum()
    total_nulls = null_summary.sum()
    
    print(f"File: {file}")
    print("Null values per column:")
    print(null_summary)
    print(f"Total null values: {total_nulls}")
    print("-" * 40)
```

```python
column = 'ref_sdcch_ass_fail_rateper'

# Convert the column to numeric if needed (handling errors)
data[column] = pd.to_numeric(data[column], errors='coerce')

# Drop rows with missing values in the column
data = data.dropna(subset=[column])

# Visualization 1: Line Plot (to see trends over index or time)
plt.figure(figsize=(10, 6))
plt.plot(data.index, data[column], marker='o', linestyle='-', color='blue', alpha=0.7)
plt.title('Behavior of ref_sdcch_ass_fail_rateper (Line Plot)', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Failure Rate (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Visualization 2: Histogram (to see the distribution)
plt.figure(figsize=(8, 5))
sns.histplot(data[column], bins=30, kde=True, color='green', alpha=0.6)
plt.title('Distribution of ref_sdcch_ass_fail_rateper', fontsize=14)
plt.xlabel('Failure Rate (%)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()

# Visualization 3: Box Plot (to see outliers and spread)
plt.figure(figsize=(8, 5))
sns.boxplot(x=data[column], color='orange', width=0.5)
plt.title('Box Plot of ref_sdcch_ass_fail_rateper', fontsize=14)
plt.xlabel('Failure Rate (%)', fontsize=12)
plt.show()

# Optional: Print basic statistics for better understanding
print("Basic Statistics of ref_sdcch_ass_fail_rateper:")
print(data[column].describe())
```
